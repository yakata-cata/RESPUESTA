```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit import execute, Aer
import numpy as np
```

```python
oraculo_b = QuantumCircuit(9)

#oraculo_b.u(0,0,0,0,1,1,1,1,)

oraculo_b.x(1)
oraculo_b.barrier()

for x in range(8):
    oraculo_b.cx(x,8)

oraculo_b.barrier()
oraculo_b.x(1)
oraculo_b.barrier()

oraculo_b.draw()
```

```python
cb=QuantumCircuit(9,8)
cb.x(8)
cb.h(8)
cb.barrier()

for x in range(8):
    cb.h(x)
cb.barrier()

cb=cb.compose(oraculo_b)
cb.barrier

for x in range(8):
    cb.h(x)
cb.barrier()

##cb.measure([0,1,2,3],[0,1,2,3])
cb.draw()
```
